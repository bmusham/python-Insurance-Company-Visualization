# Filling the missing values/Cleaning the data (3d)
We used the following strategy to fill in the missing values:
1. Filtered the data at the granularity of the vehicle body
2. For each vehicle body type, we have filled in missing values for each column as follows:
   a. For Categorical data: mode of that column at the granularity of vehicle body type
   b. For Numeric data: mean of that column at the granularity of vehicle body type

We think this method is correct as it makes sense to take basic statistical measures like mean, median and mode to fill in the missing data. These measures give a precise value taking into consideration observations from the whole data set.

For the purpose of cleaning the data, 
5 fields are categorical in nature : Gender, Age Category, Area, Vehicle Age, Vehicle Body
6 fields are numeric in nature: Credit Score, Traffic Index, Vehicle Value, Number of Claims, Claim cost and Annual Premium

```python
print(vehicle_data.columns) #printing the columns of the dataset
```

```python
# Rename some column names so as to put them into a consistent format.
vehicle_data.rename(
    columns={"pol_number":"Policy_Number","pol_eff_dt":"Effective_Date","gender":"Gender",
             "agecat":"Age_Category","date_of_birth":"Date_Of_Birth","credit_score":"Credit_Score",
             "area":"Area","traffic_index":"Traffic_Index","veh_age":"Vehicle_Age","veh_body":"Vehicle_Body",
             "veh_value":"Vehicle_Value","numclaims":"No_Of_Claims","claimcst0":"Claims_Cost",
             "annual_premium":"Annual_Premium"},
    inplace=True)
vehicle_data.head()
```

```python
vehicle_data.isnull().sum() #counts missing values in each column
```

```python
#Columns with missing values
#Categorical - Gender, Age_Category,Area,Vehicle_Body
#Numeric - Credit_Score, Traffic_Index, Vehicle_Value
cols = ['Gender', 'Age_Category', 'Credit_Score', 'Area',
       'Traffic_Index', 'Vehicle_Body', 'Vehicle_Value']
```

```python
#Put 'Date_Of_Birth' and 'Effective_Date' columns into a proper format
vehicle_data['Date_Of_Birth'] = vehicle_data['Date_Of_Birth'].transform(lambda x: x.replace('-','/'))
vehicle_data['Effective_Date'] = vehicle_data['Effective_Date'].transform(lambda x: x.replace('-','/'))
vehicle_data
```

```python
# aggregate measures across whole columns in data set
#Categorical - Gender, Age_Category,Area,Vehicle_Body
print('Aggregate measures across whole columns in data set')
print('Mode for Categorical Columns')
cat_cols = ['Gender','Age_Category','Area','Vehicle_Body']
for column in cat_cols:
    print(column+': '+ str(vehicle_data[column].mode()[0]))
print('------------------------------------------------')
#Numeric - Credit_Score, Traffic_Index, Vehicle_Value
print('Mean for Numeric Columns')
num_cols = ['Credit_Score','Traffic_Index','Vehicle_Value']
for column in num_cols:
    print(column+': '+ str(vehicle_data[column].mean()))
```

```python
#Filling missing value for Vehicle_Body
veh_body_mode = vehicle_data['Vehicle_Body'].mode()[0]
print('Mode: ',veh_body_mode)
vehicle_data['Vehicle_Body'] = vehicle_data['Vehicle_Body'].transform(lambda x: x.replace(np.nan,veh_body_mode))
vehicle_data.isnull().sum() #Checking missing values
```

```python
# aggregate measures at the level of granularity of 'Vehicle_Body'
#Filling missing values for Numeric columns- Credit_Score, Traffic_Index, Vehicle_Value
print("Aggregate measures at the level of granularity of 'Vehicle_Body'")
for col in num_cols:
    mean = vehicle_data.groupby('Vehicle_Body')[col].mean()
    print(mean)
    print('-------------------------------------------------------')
```

```python
print("Aggregate measures at the level of granularity of 'Vehicle_Body'")
#Categorical - Gender, Age_Category,Area,Vehicle_Body
veh_bodies = vehicle_data.groupby('Vehicle_Body').groups.keys()

for column in cat_cols:
    print('Mode at granlularity of Vehicle_body for ',column)
    for body in veh_bodies:
        print('Mode for '+body+' is: ',vehicle_data.groupby('Vehicle_Body').get_group(body)[column].mode()[0])
```

```python
#Filling in the missing values for Numeric columns - Credit_Score, Traffic_Index, Vehicle_Value.
for col in num_cols:
    vehicle_data[col] = vehicle_data.groupby('Vehicle_Body')[col].transform(lambda x:x.fillna(x.mean()))
```

```python
#Filling in the missing values for Categorical Columns - Gender, Age_Category,Area,Vehicle_Body
for col in cat_cols:
    vehicle_data[col] = vehicle_data.groupby('Vehicle_Body')[col].transform(lambda x:x.fillna(x.mode()[0]))
```

# Histogram

**The Histogram depicts the number of policy takers in each age group.**
It shows the Age Categories on the x-axis, in the intervals of 10 like 0-10,10-20, 20-30, etc and
Frequency of the number of policies on the y-axis

```python
#Histogram - To show number of people taking policies in different age groups
from datetime import datetime
def age(date_of_birth):
    today = datetime.today()
    age = today.year - date_of_birth.year - ((today.month, today.day) < (date_of_birth.month, date_of_birth.day))
    return age
vehicle_data['Age'] = vehicle_data['Date_Of_Birth'].transform(lambda x: age(datetime.strptime(x, '%m/%d/%Y')))
ages = vehicle_data['Age']
# setting the ranges and no. of intervals 
range = (0, 100) 
bins = 10

# plotting a histogram 
plt.hist(ages, bins, range,color = 'red', histtype = 'step', rwidth = 0.8) 

# x-axis label 
plt.xlabel('Age of the Policy Holder') 

# frequency label 
plt.ylabel('No. of people') 

# plot title 
plt.title('Number of people taking policies in different age groups') 

# function to show the plot 
plt.show() 
```

We observe that people in age groups 40-60 take the highest number of policies
The number of people taking the policies goes up until the age of 60 and then gradually comes down as the age increases
